```python
import plotly.express as px
fig = px.histogram(a, 
                   x=a['Manufacturing_year'], 
                   marginal='box', 
                   nbins=16, 
                   title='Distribution of Year')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.histogram(a, 
                   x=a['Price'], 
                   marginal='box', 
                   color='Transmission', 
                   color_discrete_sequence=['green', 'grey'], 
                   title='Annual Medical Charges')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.histogram(a, 
                   x=a['KM driven'],
                   marginal='box', 
                   color_discrete_sequence=['green', 'grey'], 
                   title='Annual Medical Charges')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.scatter(a, 
                 x=a['KM driven'],
                 y=a['Price'], 
                 opacity=0.8, 
                 title='Kilometers driven vs. Selling Price')
fig.update_traces(marker_size=5)
fig.show()
```
